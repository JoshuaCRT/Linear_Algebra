Without the use of any python libraries or modules, develop a function that can perform matrix addition, given two numpy matrices.


```python
def matrixAdd(matrix1, matrix2):
    if len(matrix1) != len(matrix2) or len(matrix1[0]) != len(matrix2[0]):
        return "Matrices must be of the same size for addition."
    
    result = []
    for i in range(len(matrix1)):
        row = []
        for j in range(len(matrix1[0])):
            row.append(matrix1[i][j] + matrix2[i][j])
        result.append(row)
    
    return result

# Example matrices
matrix_A = [[1, 2, 3],
            [4, 5, 6],
            [7, 8, 9]]

matrix_B = [[9, 8, 7],
            [6, 5, 4],
            [3, 2, 1]]

# Perform matrix addition
result_matrix = matrixAdd(matrix_A, matrix_B)
print("Result of matrix addition:")
for row in result_matrix:
    print(row)
```

Without the use of any python libraries or modules, develop a function that can multiply two numpy matrices.


```python
def matrixMult(matrix1, matrix2):
    if len(matrix1[0]) != len(matrix2):
        return "Number of columns in the first matrix must match the number of rows in the second matrix for multiplication."

    result = []
    for i in range(len(matrix1)):
        row = []
        for j in range(len(matrix2[0])):
            val = 0
            for k in range(len(matrix2)):
                val += matrix1[i][k] * matrix2[k][j]
            row.append(val)
        result.append(row)

    return result

# Example matrices
matrix_A = [[1, 2, 3],
            [4, 5, 6],
            [7, 8, 9]]

matrix_B = [[9, 8, 7],
            [6, 5, 4],
            [3, 2, 1]]

# Perform matrix multiplication
result_matrix = matrixMult(matrix_A, matrix_B)
print("Result of matrix multiplication:")
for row in result_matrix:
    print(row)
```

The rule of distributivity states that given two matrices $A$ and $B$ and a scalar, $k$, then $k(A+B)=kA+kB$. Instead of writing a proof mathematically, develop two codes for $k(A+B)$ and $kA+kB$."
   ]

```python
def scalar_multiply(matrix, k):
    result = []
    for row in matrix:
        new_row = [elem * k for elem in row]
        result.append(new_row)
    return result

def matrix_addition(matrix1, matrix2):
    if len(matrix1) != len(matrix2) or len(matrix1[0]) != len(matrix2[0]):
        return "Matrices must be of the same size for addition."
    
    result = []
    for i in range(len(matrix1)):
        row = []
        for j in range(len(matrix1[0])):
            row.append(matrix1[i][j] + matrix2[i][j])
        result.append(row)
    
    return result

def distributive_property(A, B, k):
    # Calculate k(A + B)
    A_plus_B = matrix_addition(A, B)
    k_times_A_plus_B = scalar_multiply(A_plus_B, k)
    
    # Calculate kA + kB
    kA = scalar_multiply(A, k)
    kB = scalar_multiply(B, k)
    kA_plus_kB = matrix_addition(kA, kB)
    
    return k_times_A_plus_B, kA_plus_kB

# Example matrices
matrix_A = [
    [2, 3],
    [4, 5]
]
matrix_B = [
    [1, 0],
    [2, 1]
]

# Scalar
k = 2

# Calculate k(A + B) and kA + kB
result_distributive = distributive_property(matrix_A, matrix_B, k)
print("k(A + B):")
for row in result_distributive[0]:
    print(row)

print("\nkA + kB:")
for row in result_distributive[1]:
    print(row)
```

Without using a python library or modules develop a function that can extract the diagonal of a numpy matrix.

```python
def extract_diagonal(matrix):
    if len(matrix) != len(matrix[0]):
        return "Matrix must be square for diagonal extraction."
    
    diagonal = []
    for i in range(len(matrix)):
        diagonal.append(matrix[i][i])
    
    return diagonal

# Example matrix
example_matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Extract diagonal
result_diagonal = extract_diagonal(example_matrix)
print("Diagonal elements:")
print(result_diagonal)
```

Without using a python library or modules develop a function that can find a trace of a numpy matrix.

```python
def matrix_trace(matrix):
    if len(matrix) != len(matrix[0]):
        return "Matrix must be square for trace calculation."
    
    trace = 0
    for i in range(len(matrix)):
        trace += matrix[i][i]
    
    return trace

# Example matrix
example_matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Calculate trace
result_trace = matrix_trace(example_matrix)
print("Trace of the matrix:")
print(result_trace)

```
